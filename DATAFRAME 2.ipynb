```python
import pandas as pd
```

```python
df=pd.read_csv("/Users/ISHU/Desktop/WORK/DATASETS/employees.csv",parse_dates=["Start Date","Last Login Time"])
df["Senior Management"]=df["Senior Management"].astype("bool")
df["Gender"]=df["Gender"].astype("category")
df.sort_values("First Name",inplace=True)
df.head(3)
```

```python
df[df["Gender"] == "Male"].head(3)
# df["Gender"] == "Male" here we get a boolean series of values
#df[df["Gender"] == "Male"] creates a new data frame with only males
# if we use = then it is going to assign sam evalues to whole column
```

```python
df[df["Senior Management"]].head(3) # since the series is boolean it wil automatically selct true values in new dataframe
```

```python
df[df["Team"]!= "Marketing"].head(3)  # new df with no marketing
```

### .DUPLICATED() METHOD

```python
df[df["First Name"].duplicated()].head(5)
# it is showing only those values that are duplicate and removed one value of every duplicate row
#out of 4 Aron its only showing three Aron
```

```python
df[df["First Name"].duplicated(keep=False)].head(5)
# it will mark all the values which occour more than one as Duplicate
```

```python
df[~df["First Name"].duplicated(keep=False)].head(5)
#it will only show values which occur only once in dataframe
```

## DROP_DUPLICATES()METHOD

```python
len(df)
```

```python
df.drop_duplicates()
# no duplicates are deleted because it only deletes those rows which have same same value in more than one rows across all columns
```

```python
df.drop_duplicates(subset=["First Name"],keep= "first").head(5)
# removed rows which have same First Name value in dataframe and kept first value as non duplicate
```

```python
df.drop_duplicates(subset=["First Name"],keep= "last").head(4)
#kept only one non duplicate from last 
```

```python
df.drop_duplicates(subset=["First Name"],keep= False)
#this will remove every value that is duplicate in dataframe
```

```python
df.drop_duplicates(subset=["First Name","Team"]).head(10)
```
