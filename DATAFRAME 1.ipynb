### HANDLING NULL VALUES

To change values in orginal database we hace to use inplace=True

```python
nba=pd.read_csv("/Users/ISHU/Desktop/WORK/DATASETS/nba.csv")
nba.tail(3)
```

```python
nba.dropna()  #remove all rows with NaN values
```

```python
nba.dropna(how= "all") # THIS REMOVES ROWS IF ONLY ALL VALUES IN ROWS ARE NaN
```

```python
nba.dropna(axis=1 ,how="all") #"columns" this will remove all the colums with null values
```

```python
nba.dropna(subset=["Salary"]) # all rows with NaN values removed from salary column
```

```python
nba.dropna(subset=["Salary","College"]) # all rows with NaN values removed from salary & college column
```

## RANK() METHOD

```python
nba=pd.read_csv("/Users/ISHU/Desktop/WORK/DATASETS/nba.csv").dropna(how="all")
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
nba["Salary-Rank"]=nba["Salary"].rank(ascending=False).astype(int) #rank method
nba.head(3)
```

```python
nba.sort_values(by="Salary",ascending=False).head(7)
```
